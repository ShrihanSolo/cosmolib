```python

import getdist
from getdist import plots, MCSamples
from getdist.paramnames import escapeLatex, makeList, mergeRenames
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from pathlib import Path
%matplotlib inline


g = plots.getSubplotPlotter(width_inch=8)
```

```python
fidA = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-clean-des-y3-fidA-nautilus.txt'
fidB = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-clean-des-y3-fidB-nautilus.txt'
fidC = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-clean-des-y3-fidC-nautilus.txt'
impfidA = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-clean-des-y3-psfA-nautilus.txt'
impfidB = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-clean-des-y3-psfB-nautilus.txt'
impfidC = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-clean-des-y3-psfC-nautilus.txt'

#imprintresp = Path.home() / 'weaklens' / 'cosmolib' / 'output' / 'shrihan-des-y3-response2_A.txt'
fidA_chains = np.genfromtxt(fidA, names=True)
fidB_chains = np.genfromtxt(fidB, names=True)
fidC_chains = np.genfromtxt(fidC, names=True)
impfidA_chains = np.genfromtxt(impfidA, names=True)
impfidB_chains = np.genfromtxt(impfidB, names=True)
impfidC_chains = np.genfromtxt(impfidC, names=True)
```

```python

omm = 0.#,#0.03
oss = 0.#,#0.03

uu = fidA_chains
fidA_samples = MCSamples(samples=np.c_[uu['cosmological_parametersomega_m']+omm,uu['COSMOLOGICAL_PARAMETERSSIGMA_8'],
                                (uu['COSMOLOGICAL_PARAMETERSSIGMA_8'])*np.sqrt((uu['cosmological_parametersomega_m'])/0.3)+oss,
                                 uu['cosmological_parametersn_s'], uu['cosmological_parametersomega_b'], uu['cosmological_parametersh0']],
                                 weights=np.exp(uu['log_weight']), names = ['Om','sigma8','s8','ns', 'Ob', 'h0'], labels = ['\Omega_{m}','\sigma_8','S_8','n_s','\Omega_b', 'h'])


uu = fidB_chains
fidB_samples = MCSamples(samples=np.c_[uu['cosmological_parametersomega_m']+omm,uu['COSMOLOGICAL_PARAMETERSSIGMA_8'],
                                (uu['COSMOLOGICAL_PARAMETERSSIGMA_8'])*np.sqrt((uu['cosmological_parametersomega_m'])/0.3)+oss,
                                 uu['cosmological_parametersn_s'], uu['cosmological_parametersomega_b'], uu['cosmological_parametersh0']],
                                 weights=np.exp(uu['log_weight']), names = ['Om','sigma8','s8','ns', 'Ob', 'h0'], labels = ['\Omega_{m}','\sigma_8','S_8','n_s','\Omega_b', 'h'])

uu = fidC_chains
fidC_samples = MCSamples(samples=np.c_[uu['cosmological_parametersomega_m']+omm,uu['COSMOLOGICAL_PARAMETERSSIGMA_8'],
                                (uu['COSMOLOGICAL_PARAMETERSSIGMA_8'])*np.sqrt((uu['cosmological_parametersomega_m'])/0.3)+oss,
                                 uu['cosmological_parametersn_s'], uu['cosmological_parametersomega_b'], uu['cosmological_parametersh0']],
                                 weights=np.exp(uu['log_weight']), names = ['Om','sigma8','s8','ns', 'Ob', 'h0'], labels = ['\Omega_{m}','\sigma_8','S_8','n_s','\Omega_b', 'h'])


uu = impfidA_chains
impfidA_samples = MCSamples(samples=np.c_[uu['cosmological_parametersomega_m']+omm,uu['COSMOLOGICAL_PARAMETERSSIGMA_8'],
                                (uu['COSMOLOGICAL_PARAMETERSSIGMA_8'])*np.sqrt((uu['cosmological_parametersomega_m'])/0.3)+oss,
                                 uu['cosmological_parametersn_s'], uu['cosmological_parametersomega_b'], uu['cosmological_parametersh0']],
                                 weights=np.exp(uu['log_weight']), names = ['Om','sigma8','s8','ns', 'Ob', 'h0'], labels = ['\Omega_{m}','\sigma_8','S_8','n_s','\Omega_b', 'h'])

uu = impfidB_chains
impfidB_samples = MCSamples(samples=np.c_[uu['cosmological_parametersomega_m']+omm,uu['COSMOLOGICAL_PARAMETERSSIGMA_8'],
                                (uu['COSMOLOGICAL_PARAMETERSSIGMA_8'])*np.sqrt((uu['cosmological_parametersomega_m'])/0.3)+oss,
                                 uu['cosmological_parametersn_s'], uu['cosmological_parametersomega_b'], uu['cosmological_parametersh0']],
                                 weights=np.exp(uu['log_weight']), names = ['Om','sigma8','s8','ns', 'Ob', 'h0'], labels = ['\Omega_{m}','\sigma_8','S_8','n_s','\Omega_b', 'h'])

uu = impfidC_chains
impfidC_samples = MCSamples(samples=np.c_[uu['cosmological_parametersomega_m']+omm,uu['COSMOLOGICAL_PARAMETERSSIGMA_8'],
                                (uu['COSMOLOGICAL_PARAMETERSSIGMA_8'])*np.sqrt((uu['cosmological_parametersomega_m'])/0.3)+oss,
                                 uu['cosmological_parametersn_s'], uu['cosmological_parametersomega_b'], uu['cosmological_parametersh0']],
                                 weights=np.exp(uu['log_weight']), names = ['Om','sigma8','s8','ns', 'Ob', 'h0'], labels = ['\Omega_{m}','\sigma_8','S_8','n_s','\Omega_b', 'h'])

g.settings.axes_labelsize = 20
g.settings.axes_fontsize =15
g.settings.legend_fontsize = 20
g.settings.linewidth = 2

```

```python
g.triangle_plot([fidA_samples, fidB_samples, fidC_samples, impfidA_samples, impfidB_samples, impfidC_samples], ['Om', 'sigma8', 's8'], filled=[False,False,False, True, True, True], 
                colors=['coral','coral', 'coral', 'black','black', 'black'], alphas=[1.0, 1.0, 1.0, 0.5, 0.5, 0.5],
                legend_labels=['DES Fiducial A', 'DES Fiducial B', 'DES Fiducial C', 'DES PSF Imprint A', 'DES PSF Imprint B', 'DES PSF Imprint C'],  legend_loc='upper right',fontsize=40, 
                contour_colors=['coral','coral', 'coral', 'black','black', 'black'],
                ls = ['-','-','-','-'], contour_ls = ['-','-','-','-'])
plt.savefig('des_psf_all.png', dpi=600, bbox_inches='tight', transparent=True)
```
